```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('fuel.csv')
```

```python
data
```

```python
data.head()
```

```python
data = data[["ENGINESIZE" , "CO2EMISSIONS"]]
```

```python
data
```

```python
plt.scatter(data["ENGINESIZE"], data["CO2EMISSIONS"], color="blue")
plt.xlabel("ENGINESIZE")
plt.ylabel("CO2EMISSIONS")
```

```python
len(data)
```

```python
train = data[: (int(len(data)*0.8))]
```

```python
len(train)
```

```python
test =  data[(int(len(data)*0.8)) :]
```

```python
len(test)
```

```python
regr = linear_model.LinearRegression()
train_x = np.array(train[["ENGINESIZE"]])
train_y = np.array(train[["CO2EMISSIONS"]])
regr.fit(train_x,train_y)
```

```python
print("B0 :", regr.coef_)
print("B1 :", regr.intercept_)
B0 = regr.intercept_
B1 = regr.coef_
```

```python
plt.scatter(data["ENGINESIZE"], data["CO2EMISSIONS"], color="blue")
plt.xlabel("ENGINESIZE")
plt.ylabel("CO2EMISSIONS")
plt.plot(train_x, B0 + B1 *train_x, '-r')
```

```python
def get_regression_predictions(input_feature, B0 , B1):
    prediction = B0 + B1 * input_feature
    return prediction
```

```python
engine_size = 3.5
estimated_co2Emission  = get_regression_predictions(engine_size,B0,B1)
print("Estimated_c2emission : " , estimated_co2Emission)
```

```python
test_x = np.array(test[["ENGINESIZE"]])
test_y = np.array(test[["CO2EMISSIONS"]])
test_yy = regr.predict(test_x)
```

```python
Error  = np.mean((test_y - test_yy) ** 2)
Error
```
